### Código

```python
from sklearn.metrics import accuracy_score
```

```python
import numpy as np

def euclidean_distance(point1, point2):
    """Calcula la distancia euclidiana entre dos puntos."""
    squared_distance = np.sum((point1 - point2) ** 2)
    return np.sqrt(squared_distance)

def knn_classifier(X_train, y_train, test_instance, k):
    """Clasifica una instancia de prueba utilizando el clasificador KNN."""
    distances = []
    for i in range(len(X_train)):
        distance = euclidean_distance(X_train[i], test_instance)
        distances.append((distance, y_train[i]))
    distances.sort()
    neighbors = distances[:k]
    class_counts = {}
    for _, label in neighbors:
        class_counts[label] = class_counts.get(label, 0) + 1
    sorted_class_counts = sorted(class_counts.items(), key=lambda x: x[1], reverse=True)
    return sorted_class_counts[0][0]
```

```python
# Cargar el banco de datos Iris
from sklearn.datasets import load_iris
iris = load_iris()
X = iris.data
y = iris.target
```

```python
def leave_one_out(X, y, k=3):

    y_true = []
    y_pred = []
    n_samples = X.shape[0]
    error = 0

    for i in range(n_samples):
        X_train = np.delete(X, i, axis=0)
        y_train = np.delete(y, i)

        X_test = X[i]
        y_test = y[i]

        y_true.append(y[i])

        predicted_class = knn_classifier(X_train, y_train, X_test, k)
        y_pred.append(predicted_class)
        if predicted_class != y_test:
            error += 1

    error /= n_samples

    return error,y_true,y_pred
```

```python
def hold_out_a(X, y, test_size=0.2, k=3):
    n_samples = X.shape[0]
    n_test = int(test_size * n_samples)

    # Shuffle data
    indices = np.random.permutation(n_samples)
    X_shuffled = X[indices]
    y_shuffled = y[indices]

    X_train = X_shuffled[:-n_test]
    y_train = y_shuffled[:-n_test]

    X_test = X_shuffled[-n_test:]
    y_test = y_shuffled[-n_test:]

    error = 0
    for i in range(len(X_test)):
        predicted_class = knn_classifier(X_train, y_train, X_test[i], k)          
        if predicted_class != y_test[i]:
            error += 1

    error /= len(X_test)
    return error
```

```python
def hold_out(X, y, test_size=0.2, k=3):
    n_samples = X.shape[0]
    n_test = int(test_size * n_samples)
    y_pred = []

    # Shuffle data
    indices = np.random.permutation(n_samples)
    X_shuffled = X[indices]
    y_shuffled = y[indices]

    X_train = X_shuffled[:-n_test]
    y_train = y_shuffled[:-n_test]

    X_test = X_shuffled[-n_test:]
    y_test = y_shuffled[-n_test:]

    error = 0
    for i in range(len(X_test)):
        predicted_class = knn_classifier(X_train, y_train, X_test[i], k)
        y_pred.append(predicted_class)
        if predicted_class != y_test[i]:
            error += 1

    error /= len(X_test)
    return error,y_test,y_pred
```

```python
def k_fold_cross_validation(X, y, k=10):
    n_samples = X.shape[0]
    fold_size = n_samples // k
    error = 0

    for i in range(k):
        start = i * fold_size
        end = (i + 1) * fold_size

        X_test = X[start:end]
        y_test = y[start:end]

        X_train = np.concatenate((X[:start], X[end:]), axis=0)
        y_train = np.concatenate((y[:start], y[end:]), axis=0)

        error = hold_out_a(X_train, y_train, test_size=0.2)

    error /= k
    return error
```

```python
# Calcular el error de sustitución usando Hold-Out
error_hold_out_iris,y_test_HO,y_pred_HO = hold_out(X, y)
print("Error de sustitución (Hold-Out):", error_hold_out_iris)
accuracy_HO_iris = accuracy_score(y_test_HO, y_pred_HO)
print("Accuracy: %.2f%%" % (accuracy_HO_iris*100.0))

# Calcular el error de sustitución usando Leave-One-Out
error_leave_one_out_iris, y_test_LO, y_pred_LO = hold_out(X, y)
print("Error de sustitución (Leave-One-Out):", {error_leave_one_out_iris})
accuracy_LO_iris = accuracy_score(y_test_LO, y_pred_LO)
print("Accuracy: %.2f%%" % (accuracy_LO_iris*100.0))

#Calcular el error de sustitucion usando 10-Fold Cross-validation
error_cross_val_iris= k_fold_cross_validation(X, y)
print("Error de sustitución (10-Fold Cross-validation):", error_cross_val_iris)
```
